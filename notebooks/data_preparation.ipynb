```python
# Remove this file
#files.remove(os.path.basename(__file__))

# Remove hidden files
files_old = files
for filename in files_old:
    if filename[0] == '.':
        files.remove(filename)
        
# Remove directories
#files_new = []
#for i in range(len(files)):
#    if not os.path.isdir(files[i]):
#        files_new.append(files[i])
        
#files = files_new
files = sorted(files)

# Convert each file to png and save it
for file in files:
    images = convert_from_path(f'../raw_data/pdf/{file}', 50, grayscale=True, size=(1024,None))
    
    # This is a list, so extract just the image
    images = images[0]
    
    # Remove extension
    filename = file.split('.')[0]
    images.save(filename + '.png')
    
    # Delete original file
    #os.remove(file)
```
